```python
# Importing the Libraries
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
from scipy import stats
import tensorflow as tf
import seaborn as sns
from pylab import rcParams
from sklearn import metrics

random_seed = 42
```

```python
# Retrieving the Dataset. The Dataset used is from Wireless Sensor Data Mining Lab(WISDM)

columns = ['user', 'activity', 'timestamp', 'x-axis', 'y-axis', 'z-axis']
dataset = pd.read_csv('WISDM_ar_v1.1_raw.txt', header = None, names = columns)
dataset.head()
dataset = dataset.dropna()

#Here, we will try to predict the time of activity based on the values of the three axis.
```

```python
#Visualizing the axis values using matplotlib

def activity_visual(activity, dataset):
  data = dataset[dataset['activity'] == activity][['x-axis', 'y-axis', 'z-axis']][:200]
  axis = data.plot(subplots = True, figsize = (16,12), title = activity)
  
  for ax in axis:
    ax.legend(loc = 'lower left', bbox_to_anchor = (1.0,0.5))

#This function takes the activity as input from the dataset and displays the reading for 200 timesteps.
```

```python
# Data Preprocessing. the inputs are the values of the three axes and the output is the activity.

# the input will be in the form of [samples, timesteps, features].

#Here, features  = 3 (Xaxis, yaxis and zaxis)

n_timesteps = 200
n_features = 3
steps = 20
segments = []
labels = []
for i in range(0, len(dataset) - n_timesteps, steps):
  x_segment = dataset['x-axis'].values[i: i + n_timesteps]
  y_segment = dataset['y-axis'].values[i: i + n_timesteps]
  z_segment = dataset['z-axis'].values[i: i + n_timesteps]
  label = stats.mode(dataset['activity'][i: i + n_timesteps])[0][0]
  segments.append([x_segment, y_segment, z_segment])
  labels.append(label)
```

```python
np.array(segments).shape # we convert the list into an array and check the shape
```

```python
labels[0]
```

```python
# Since the shape is in the form [samples,features, timesteps], 

#we convert that into [samples,timesteps,features]

reshaped_segments = np.asarray(segments, dtype = np.float32).reshape(-1, n_timesteps, n_features)
labels = np.asarray(pd.get_dummies(labels), dtype = np.float32)

#Here we apply one-hot encoding to the activity labels, since neural networks take in only numeric values.
#this is acheieved by the get_dummies() method
```

```python
labels[0]
```

```python
# We split the dataset into training set and test set. 80% training set and 20% test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(reshaped_segments, labels, test_size = 0.2, random_state = random_seed)
```

```python
# Building the Model - 2 LSTM layers stacked onto each other with 64 neurons in each layer.

n_classes = 6 #number of activities which being classified
n_neurons = 64 #number of neurons in each layer

def lstm_model(inputs):
  
  # First we define the variables for weights and biases to be updated during backpropagation.
  # we randomly generate the value for weights and biases which will then be tuned during backprop.
  W = {
      'hidden' : tf.Variable(tf.random_normal([n_features, n_neurons])),
      'output' : tf.Variable(tf.random_normal([n_neurons, n_classes]))  
  }
  
  B = {
      'hidden' : tf.Variable(tf.random_normal([n_neurons], mean = 1.0)),
      'output' : tf.Variable(tf.random_normal([n_classes]))  
  }
  
  X = tf.transpose(inputs, [1, 0, 2])
  X = tf.reshape(X, [-1, n_features])
  hidden = tf.nn.relu(tf.matmul(X, W['hidden']) + B['hidden']) # we use relu as the activation function for the intermediate layers.
  hidden = tf.split(hidden, n_timesteps, 0) # we split our tensor into sub tensors.
  
  # we then add the lstm layers 
  
  lstm_layer = [tf.contrib.rnn.BasicLSTMCell(n_neurons, forget_bias = 1.0) for _ in range(2)]
  lstm_layer = tf.contrib.rnn.MultiRNNCell(lstm_layer)
  
  
  #we then add the output layer
  
  output_layer, _ = tf.contrib.rnn.static_rnn(lstm_layer, hidden, dtype = tf.float32)
  
  #we then retrieve the output for the last timestep
  
  last_output = output_layer[-1]
  
  return tf.matmul(last_output, W['output']) + B['output']
```

```python
# Inorder to add the input and get the output from the model, we have to create placeholders.
# Placeholders are lists that hold the input and output values.
tf. reset_default_graph() 
X = tf.placeholder(tf.float32, [None, n_timesteps, n_features], name = "input")# providing a name for the input tensor
Y = tf.placeholder(tf.float32, [None, n_classes])
```

```python
y_pred = lstm_model(X)

active_pred = tf.nn.softmax(y_pred, name = 'y_') # providing a name to the output tensor
```

```python
# loss function

l2_loss = 0.0015

l2 = l2_loss * \
      sum(tf.nn.l2_loss(tf_var) for tf_var in tf.trainable_variables())

loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits = y_pred, labels = Y)) + l2
```

```python
learning_rate = 0.0025

optimizer = tf.train.AdamOptimizer(learning_rate = learning_rate).minimize(loss)

correct_pred = tf.equal(tf.argmax(active_pred, 1), tf.argmax(Y, 1))
accuracy = tf.reduce_mean(tf.cast(correct_pred, dtype = tf.float32))
```

```python
# Training the Model

epochs = 50
batch_size = 1024
saver = tf.train.Saver()

history = dict(train_loss=[], 
                     train_acc=[], 
                     test_loss=[], 
                     test_acc=[])

sess=tf.InteractiveSession()
sess.run(tf.global_variables_initializer())

train_count = len(X_train)

for i in range(1, epochs + 1):
    for start, end in zip(range(0, train_count, batch_size),
                          range(batch_size, train_count + 1,batch_size)):
        sess.run(optimizer, feed_dict={X: X_train[start:end],
                                       Y: y_train[start:end]})

    _, acc_train, loss_train = sess.run([active_pred, accuracy, loss], feed_dict={
                                            X: X_train, Y: y_train})

    _, acc_test, loss_test = sess.run([active_pred, accuracy, loss], feed_dict={
                                            X: X_test, Y: y_test})

    history['train_loss'].append(loss_train)
    history['train_acc'].append(acc_train)
    history['test_loss'].append(loss_test)
    history['test_acc'].append(acc_test)

    if i != 1 and i % 10 != 0:
        continue

    print(f'epoch: {i} test accuracy: {acc_test} loss: {loss_test}')
    
predictions, acc_final, loss_final = sess.run([active_pred, accuracy, loss], feed_dict={X: X_test, Y: y_test})

print()
print(f'final results: accuracy: {acc_final} loss: {loss_final}')
```

```python
#we then export this model to disk into two files - lstm.pbtxt and lstm.ckpt

pickle.dump(predictions, open("predictions.p", "wb"))
pickle.dump(history, open("history.p", "wb"))
tf.train.write_graph(sess.graph_def, 'content','lstm.pbtxt')
saver.save(sess, save_path = 'content/lstm.ckpt')
sess.close()
```

```python
# Plotting the graph to check the accuracy and loss for training and test phase

plt.figure(figsize=(12, 8))

plt.plot(np.array(history['train_loss']), "r--", label="Train loss")
plt.plot(np.array(history['train_acc']), "g--", label="Train accuracy")

plt.plot(np.array(history['test_loss']), "r-", label="Test loss")
plt.plot(np.array(history['test_acc']), "g-", label="Test accuracy")

plt.title("Training session's progress over iterations")
plt.legend(loc='upper right', shadow=True)
plt.ylabel('Training Progress (Loss or Accuracy values)')
plt.xlabel('Training Epoch')
plt.ylim(0)

plt.show()
```

```python
LABELS = ['Downstairs', 'Jogging', 'Sitting', 'Standing', 'Upstairs', 'Walking']
```

```python
# Creating the confusion matrix to check the performace of the model for a particular activity

max_test = np.argmax(y_test, axis=1)
max_predictions = np.argmax(predictions, axis=1)
confusion_matrix = metrics.confusion_matrix(max_test, max_predictions)

plt.figure(figsize=(16, 14))
sns.heatmap(confusion_matrix, xticklabels=LABELS, yticklabels=LABELS, annot=True, fmt="d");
plt.title("Confusion matrix")
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show();
```
